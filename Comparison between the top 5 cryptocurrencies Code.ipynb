```python
#import the required libraries

import pandas as pd
import numpy as np
```

```python
#import the raw data

Ctypto_coins = pd.read_csv('crypto_coins.csv')
```

```python
#print the the data set in order to understand its variables

Ctypto_coins
```

```python
#Count the number of variables & observations

print('No. of observations & No. of variables:', Ctypto_coins.shape)
```

```python
#Count the number of Null values

Ctypto_coins.isna().sum()
```

```python
#Check the Dtype

Ctypto_coins.info()
```

```python
#Drop the "Timestamp" variable as it's irrelevant

Ctypto_coins.drop(['Timestamp'], axis = 1, inplace = True)
```

```python
#Replace the NaN values with 0

Ctypto_coins = Ctypto_coins.fillna(0)
```

```python
Ctypto_coins
```

```python
# I wanted the drop the date column just for a while till I round all the columns then int64 

new_df = Ctypto_coins.iloc[:, 1:127]
```

```python
new_df
```

```python
new_df = round(new_df).astype(int)
```

```python
new_df
```

```python
#Concat the date variable back to the data frame

new_df['Date'] = Ctypto_coins['Date']
```

```python
new_df
```

```python
#Make the date variable to be Date

new_df['Date'] = pd.to_datetime(new_df['Date']).dt.date
```

```python
new_df
```

```python
#Convert the DataFrame into csv file

new_df.to_csv('Different_coins.csv', index = False)
```
