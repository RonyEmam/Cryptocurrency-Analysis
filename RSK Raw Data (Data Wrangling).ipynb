## So in this section I have the data I have downloaded and now I need to clean this data a little before appending it to the other data set I have collected from CoinMarketCap

```python
#import the only library needed here since I will do a data wrangling only:

import pandas as pd
```

```python
#Import the raw data

df = pd.read_csv('BALN-Copy1.csv')
df
```

```python
#Conver the date column into date
df['Date'] = pd.to_datetime(df['date']).dt.date
```

```python
#Make sure that the date column is indexed

df['Date'].index = df['date']
```

```python
#Dropped the old one
df.drop(['date'], axis = 1, inplace = True)
```

```python
df
```

```python
df['Date'] = df['Date'].astype('datetime64[ns]')
```

## In the next section, I only want to retrieve the data from the range "2021.01.01 til 2022.06.28"

```python
newdata = df[df['Date'] <= '2022-06-28'] # filtering the dataframe 
print(newdata)
```

```python
dd = newdata[newdata['Date'] > '2020-12-31'] # filtering the dataframe 
print(dd)
```

```python
#Check the new edited data frame

dd
```

```python
#Get rid of the comma between the numbers in the data set

dd['balance'] = dd['balance'].replace({'RBTC':''}, regex = True)
```

```python
dd
```

```python
# I noticed that there's a mis-matching indexing so I have reset the indexes

dd = dd.reset_index(drop=True)
```

```python
dd
```

```python
dd.info()
```

```python
#Convert the columns into int64 except the date column

dd['balance'] = dd['balance'].apply(pd.to_numeric,errors='ignore')
```

```python
dd.info()
```

```python
dd['balance'] = round(dd['balance']).astype(int)
```

```python
dd.info()
```

```python
dd.head(30)
```

## This's the final data frame that has the exact data I need so I check it one more time then convert it to csv file

```python
c = dd.groupby('Date')['balance'].sum()
```

```python
c = pd.DataFrame(c)
```

```python
c
```

```python
c.to_csv(r'RSKBalanceData.csv')
```
