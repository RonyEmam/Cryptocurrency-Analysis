```python
#import the required libraries

import pandas as pd
import numpy as np
```

```python
#import the required libraries 

info = pd.read_csv('RSKBalanceData.csv')
info
```

```python
#Check the shape of the data frame

print('No. of observations & No. of variables:', info.shape)
```

```python
#Reverse the Date column so I can merge it later on with the data frame from Coinmarketcap
info = info.iloc[::-1]
```

```python
info
```

```python
#Here is the data set that we need to combine it with
data = pd.read_csv('Sunday.csv')
data
```

```python
#Get rid of the unnecessary variables

data.drop(['Unnamed: 0', 'No. of BTC Traded per month', 'circulating supply', 'Total Value Locked (TVL)', 'Close_MA',
          'Close_cumulative_MA', 'Close_exponential_MA'], axis = 1, inplace = True)
```

```python
#check the final results
data
```

```python
#Here I merge the two previous presented data frames, with condition that the two dates are equal,
#so I have excluded concat() & append(), and instead I choose merge which is equivelent to "Outer join"
df = pd.merge(data, info, on='Date', how='left')
```

```python
#Check the full new data frame
df
```

```python
#Count the NaN values in the new data set

df.isna().sum()
```

```python
#Highlight that the balance is in RBTC not USD yet, then round the variable and converted in int64 and finally
#make sure that the NaN values are replaced with 0

df['Balance_in_RBTC'] = df['balance'].replace(np.nan, 0).astype(int)
```

```python
df
```

```python
#Drop "balance column"

df.drop('balance', axis = 1, inplace = True)
```

```python
#Check what we have so far

df
```

# To get the value of Circulating Supply:


## Since Market Cap = Close Price * Circulating Supply 

```python
#Calcualting the circulating supply variable

df['Circulating Supply'] = df['Market Cap'] / df['Close']
```

```python
#Round the new column

df['Circulating Supply'] = round(df['Circulating Supply']).astype(int)
```

```python
#Make sure that the dtypes of columns are exactly as I want to be

df.info()
```

```python
#Convert the "Balance" from RBTC into USD, I googled to get how much 1 RBTC is equals to US Dollars

df['Balance in USD'] = round(df['Balance_in_RBTC'] * 18924.50).astype(int)
```

# To get the value of TVL:


## Total Value Locked (TVL) BTC - USD  = Balance in USD - Circulating Supply

```python
#Print the TVL values and rounded them then dtype to int64

df['Total Value Locked (TVL) in USD'] = round(df['Balance in USD'] - df['Circulating Supply']).astype(int)
```

```python
df
```

```python
df.isna().sum()
```

```python
#Convert the data frame into csv file
df.to_csv('TVL in USD.csv', index = False)
```
